**Задача 1**

Напишите запрос,чтобы получить все названия продуктов и соответствующие им торговые марки (brand)

```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT p.product_name, b.brand_name
        FROM public.products AS p
        JOIN public.brands AS b ON p.brand_id = b.brand_id;
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```

**Задача 2**

Напишите запрос, чтобы найти всех активных сотрудников и наименования магазинов, в которых они работают.

```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT stf.first_name, stf.last_name, str.store_name
        FROM public.staffs AS stf
        JOIN public.stores AS str ON stf.store_id = str.store_id
        WHERE stf.active = 1;
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```

**Задача 3**

Напишите запрос, чтобы перечислить всех покупателей выбранного магазина с указанием их полных имен, электронной почты и номера телефона.


```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT DISTINCT c.first_name, c.last_name, c.email, c.phone
        FROM public.customers AS c
        JOIN public.orders AS o ON c.customer_id = o.customer_id
        JOIN public.stores AS s ON o.store_id = s.store_id
        WHERE s.store_name = 'Santa Cruz Bikes';
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```

**Задача 4**

Напишите запрос для подсчета количества продуктов в каждой категории.

```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT c.category_name, COUNT(p.product_id)
        FROM public.products AS p
        JOIN public.categories AS c ON p.category_id = c.category_id
        GROUP BY c.category_name;
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```

**Задача 5**

Напишите запрос, чтобы указать общее количество заказов для каждого клиента.


```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT c.first_name, c.last_name, COUNT(o.order_id)
        FROM public.customers AS c
        JOIN public.orders AS o ON c.customer_id = o.customer_id
        GROUP BY c.customer_id
        -- Используем DESC для сортировки по убыванию
        ORDER BY COUNT(o.order_id) DESC;
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```

**Задача 6**

Напишите запрос, в котором будет указана информация о полном имени и общем количестве заказов клиентов, которые хотя бы 1 раз сделали заказ.

```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="",
        host="127.0.0.1",
        port="5432",
        options="-c client_encoding=utf8",
    )
    cursor = connection.cursor()
    cursor.execute(
        """
        SELECT c.first_name, c.last_name, COUNT(o.order_id)
        FROM public.customers AS c
        JOIN public.orders AS o ON c.customer_id = o.customer_id
        GROUP BY c.customer_id
        HAVING COUNT(o.order_id) >= 1
        ORDER BY COUNT(o.order_id);
        """
    )

    for record in cursor.fetchall():
        print(record)

except:
    print("Возникла ошибка при работе с Postgres")
```
